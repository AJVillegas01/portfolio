```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

lista_x = []
año= 1920
while año < 2000:
    
    lista_x.append(año)
    año = año + 10
    print(año)
    
print(lista_x)    
```

```python
lista_y=[106.46,123.08,132.12,152.27,180.67,205.05,227.23,249.46]
```

```python
x=np.array(lista_x)
y=np.array(lista_y)
```

```python
anno=2000
coef = np.polyfit(x,y,2) #en este caso 2 seria el grado del polinomio que queremos obtener que sera una paarabola y coef nos devolvera los coeficientes a,b y c
p= np.polyval(coef,anno)
print(f"Para el anno {anno} la prediccion es de {p}")
"""debemos de poder saber que grado hay que aplicar por eso haremos la siguiente prueba dando valores desde 0 al 9"""
```

```python
def fx (x1, coef):
    fx = 0
    n = len(coef)-1
    for p in coef:
        fx = fx + p*x1**n
        n = n -1
    return fx
```

```python
anno = 2000
for i in range(0,10):
    coef = np.polyfit(x,y,i)
    p = np.polyval(coef, anno)
    
    print (f"para el grado {i} la prediccion es {p}")
    x1 = np.linspace(1920, anno + 1, 1000)
    y1 = fx(x1, coef) # aqui tenemos la funcion que vamos a representar a continuacion con plot
    plt.figure(figsize=[20,10])
    plt.title("Cantidad de litros vs año. Para grado " + str(i))
    
    plt.scatter(x,y,s=120, c='blueviolet')
    plt.plot(x1,y1,"--",linewidth=3, color='orange')
    plt.scatter(anno, p, s=200, c='red')
    plt.yticks(range(100,320,20))
    plt.grid("on")
    ax=plt.gca()
    ax.set_xlabel("$años$")
    ax.set_ylabel("$Cantidad de litros$")
    #plt.savefig("img" + str(i)+" .jpg", dpi=600)
    plt.show()
```

```python
"A continuacion vamos a calcular el error que tiene hacer la grafica con cada grado"

anno=2000
grado = np.arange(0,30+1,1) # de 0 a 100 de 1 en 1
aproxi= np.array([])
y_pred_vec = np.array([])
for i in grado: 
    coef = np.polyfit(x,y,i)
    p= np.polyval(coef, anno)
    aproxi = np.append(aproxi,p)
    
    #para la MSE
    y_pred_vec = np.array([])
    for j in x: # para la MSE
        y_pred = np.polyval(coef,j)
        y_pred_vec = np.append(y_pred_vec, y_pred)
    #print(f"los y: {y}")
    #print(f"los y_prox: {y_pred_vec}")
    MSE = (sum( (y - y_pred_vec)**2))/len(y)
    print(f"Para grado{i} El MES es: {MSE}")
plt.figure(figsize=[20,10])
plt.title("Grado del polinomio vs predicción")
plt.plot(grado, aproxi, "--", linewidth= 3, color="red")
plt.grid("on")
plt.show()
```
